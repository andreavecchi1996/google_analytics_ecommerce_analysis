```python
from google.cloud import bigquery
import pandas as pd

# Initialize BigQuery client
client = bigquery.Client(project="sql-sandbox-435415") 
#Project ID ensures that the code is reliable and portable, 
#as it explicitly tells BigQuery which project to use, allowing it to run consistently
#across different environments and for different users.

# Example query
query = """
SELECT
  *
FROM
  `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`


LIMIT 10
"""

# Run the query and convert results to DataFrame
df = client.query(query).to_dataframe()
df
```

```python
query = """
SELECT
  count(*) as number_of_rows #count the number of rows
FROM
  `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
where
  totals.pageviews is not null #count the number of rows where 
"""

df = client.query(query).to_dataframe()
df
```
