```python
from google.cloud import bigquery
import pandas as pd

# Initialize BigQuery client
client = bigquery.Client(project="sql-sandbox-435415") 
#Project ID ensures that the code is reliable and portable, 
#as it explicitly tells BigQuery which project to use, allowing it to run consistently
#across different environments and for different users.

# Example query
query = """
SELECT
  *
FROM
  `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`


LIMIT 10
"""

# Run the query and convert results to DataFrame
df = client.query(query).to_dataframe()
df
```

This first query is used to connect to the Google Analytics database (in BigQuery) and preview the first 10 rows to understand what kind of data it contains.

```python
query = """
SELECT
  count(*) as number_of_rows #count the number of rows
FROM
  `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
where
  totals.pageviews is not null 
"""

df = client.query(query).to_dataframe()
print (df)
```

This query counts how many sessions occurred on August 1st, 2017, specifically where totals.pageviews is not null (i.e., sessions where at least one page was viewed).

There were 2,556 sessions with at least one pageview.

Running the same query without the WHERE condition gives the same result — this means all sessions on that day included at least one pageview.

```python
query = """
SELECT
  distinct fullVisitorId as number_of_visitors #count the number of rows
FROM
  `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
"""

df = client.query(query).to_dataframe()
print (df)
```

Number of unique visitor in that day: 2293

Top Traffic Sources:
What to analyze:
Which sources (e.g., google, direct, referral) brought the most visitors?

Why:
To understand where users are coming from and which marketing channels perform best.

```python
pd.set_option("display.max_rows", None)

query = """
SELECT DISTINCT trafficSource.source AS traffic_source
FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
limit 35
  """

df = client.query(query).to_dataframe()
print (df)
```

This query displays the traffic source where the customers arrived inside the eccomerce

```python
query = """
SELECT
  trafficSource.medium AS traffic_channel,
  COUNT(*) AS sessions
FROM
  `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
GROUP BY
  traffic_channel
ORDER BY
  sessions DESC
LIMIT 10
"""

df = client.query(query).to_dataframe()
print (df)
```

- (none): Likely represents direct traffic (people typing the URL directly, using bookmarks, or missing referrer data).
- referral: Visitors who came from other websites via links.
- affiliate: Visitors referred by affiliate partners.
- cpm: Traffic from cost-per-thousand-impressions ad campaigns (likely display ads).
- organic: Search engine traffic (e.g., Google, Bing).
